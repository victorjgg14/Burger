```python
import pandas as pd
import matplotlib.pyplot as plt

url1 = 'madridsecreto.csv'
url2 = 'madridelle.csv'
url3 = 'eldiarioesmadrid.csv'
url4 = 'barcelonasecreto.csv'

df_ms = pd.read_csv(url1)
df_me = pd.read_csv(url2)
df_eldm = pd.read_csv(url3)
df_bcn = pd.read_csv(url4)
```

```python
# media precio hamburguesa en madridsecreto

media_ms = df_ms.prize.mean()

media_ms
```

```python
# media el diario madrid

media_eldiariomadrid = df_eldm.prize.mean()

media_eldiariomadrid
```

```python
# media precio hamburguesa en madrid

(media_ms + media_eldiariomadrid)/2 
```

```python
df_madrid = pd.concat([df_ms, df_eldm])

df_madrid = df_madrid.reset_index()

df_madrid = df_madrid.drop(22)

df_madrid.reset_index()
```

```python
df_madbcn = pd.concat([df_madrid, df_bcn])
```

```python
df_madbcn['city'] = ['madrid','madrid','madrid','madrid','madrid','madrid','madrid','madrid','madrid','madrid',
                        'madrid','madrid','madrid','madrid','madrid','madrid','madrid','madrid','madrid','madrid',
                        'madrid','madrid','madrid','madrid','madrid','madrid','madrid','madrid','madrid','madrid',
                        'barcelona','barcelona','barcelona','barcelona','barcelona','barcelona','barcelona','barcelona',
                        'barcelona','barcelona','barcelona','barcelona','barcelona','barcelona','barcelona','barcelona']
```

```python
df_madbcn.reset_index()
```

```python
print('Precio maximo burger madrid: ',df_madrid.max())
print('Precio maximo burger bcn: ', df_bcn.max())
```

```python
meanminmax = df_madbcn.groupby('city').agg({'prize': ['mean','min', 'max']})

meanminmax
```

```python
meanminmax.T.round(2)
```

```python
meanminmax.plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Prize')
plt.title('Prize Comparison between Barcelona and Madrid')
plt.tick_params(axis='x', rotation=0)
plt.show()
```
