```python
import requests
```

```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://www.eldiario.es/guia-ocio/diez-hamburguesas-premiadas-madrid-fracasada-atun-internacional-gourmets-smash_1_10176292.html'

response = requests.get(url)

response
```

```python
response.headers
```

```python
sopa = BeautifulSoup(response.content, 'html.parser')
```

```python
sopa.prettify
```

```python
name_burger = sopa.find_all('h2')

name_burger
```

```python
lst_name = []

name_results = sopa.find_all('h2')

for e in name_results:
    
    lst_name.append(e.text)
```

```python
lst_name
```

```python
# elimino el primer y ultimo elemento de la lista porque no los quiero

lst_name.pop(0)

lst_name.remove('Qué hacer en Madrid')

lst_name
```

```python
location_burger = sopa.find_all('li')[109].text

location_burger
```

```python
lst_location = []

location_results = sopa.find_all('li')

for e in location_results:
    
    lst_location.append(e.text)
        
lst_location
```

```python
len(lst_location)
```

```python
lst_location2 = []

for e in lst_location:
    
    if e.startswith('Dónde:'):
        
        lst_location2.append(e)
        
lst_location2
```

```python
lst_location = lst_location2

lst_location
```

```python
ola = []

for e in lst_location:
    
    ola.append(e.replace('Dónde: ', '').replace('Dónde:\xa0 ', '').replace(' (Metro Príncipe Pío L6, L10 y Ramal Ópera-Príncipe Pío)', '')
              .replace(' (Metro Bilbao, L1 y L4)', '').replace(' (Metro Alonso Cano, L7)', '').replace(' (Metro Sol L1, L2 y L3)', '')
              .replace(' (Metro Arturo Soria, L4)', '').replace(' (Metro Alonso Cano, L7)', '').replace(' (Metro Bilbao, L1 y L4)', '')
              .replace(' (Metro Príncipe de Vergara, L2 y L9)', '').replace(' (Metro Pío XII, L9)', '').replace(' (Coslada)', ''))
    
ola
```

```python
lst_location = ola

lst_location
```

```python
lst_prize = []

prizes = sopa.find_all('li')

for e in prizes:
    
    lst_prize.append(e.text)
        
lst_prize
```

```python
lst_prize2 = []

for e in lst_prize:
    
    if e.startswith('Cuánto: '):
        
        lst_prize2.append(e)
        
lst_prize2
```

```python
lst_prize = lst_prize2
```

```python
alo = []

for e in lst_prize:
    
    alo.append(e.replace('Cuánto: ','').replace(',', '.').replace('€',''))

alo   
```

```python
# sobreescribo el nombre de la lista y para que al crear el df no me de error por tener la columna distinta longitud introduzco los Nan a mano

lst_prize = alo

lst_prize = ['15.90', '16', 'NaN', '13.20', '15.99', '12.50', '12.99', '18', 'NaN', '15.50']

lst_prize
```

```python
dictio = {'name': lst_name,
         'prize': lst_prize,
         'location': lst_location}

df = pd.DataFrame(dictio)
```

```python
df
```

```python
df.prize = df.prize.astype(float)
df.dtypes
```

```python
# este nulo lo relleno a sabiendas, tenia el precio scrapeado de otra pagina

df.prize[2] = 12.90
```

```python
df
```

```python
# relleno el ultimo nulo con la media de los precios ya que no tenia ese dato

df.prize = df.prize.fillna(df.prize.mean())
```

```python
df
```

```python
df.prize = df.prize.round(2)
```

```python
df
```

```python
df.to_csv('eldiarioesmadrid.csv', index=False)
```
