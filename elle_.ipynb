```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
url = 'https://www.elle.com/es/gourmet/donde-comer/g39071094/mejores-hamburgueserias-madrid/'
```

```python
response = requests.get(url)

response
```

```python
sopa = BeautifulSoup(response.content, 'html.parser')
```

```python
sopa.prettify
```

```python
name_burger = sopa.find_all('h2')

name_burger
```

```python
lst_burger = []

for e in name_burger:
    lst_burger.append(e.text)
    
lst_burger
```

```python
lst_burger
```

```python
location_burger = sopa.find_all('p')

location_burger
```

```python
loca = []

for e in location_burger:
    
    loca.append((e.text))
    
loca
```

```python
lst_location = loca
lst = []
for e in lst_location:
    
    if e.startswith('C/'):
        
        lst.append(e)
        
lst_location = lst

lst_location
```

```python
a = []
for e in lst_location:
    
    a.append(e.split('Web'))

a
```

```python
lst_location = a

lst_location
```

```python
sol = []

for e in lst_location:
    
    sol.append((e[0]))

sol
```

```python
lst_location2 = []

for e in lst_location:
    
    lst_location2.extend(e)
    
lst_location2
```

```python
lst_location = sol

lst_location
```

```python
a = [e.replace('.','') for e in lst_location]
```

```python
lst_location = a

len(lst_location)
```

```python
lst_burger
```

```python
dif_longitud = len(lst_burger) - len(lst_location)
lst_location += [None] * dif_longitud

len(lst_location)
```

```python
dictio = {'name': lst_burger,
         'location': lst_location}

df = pd.DataFrame(dictio)

df
```

```python
df.to_csv('madridelle.csv', index=False)
```
